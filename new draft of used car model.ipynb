```python
def mileage_convert(x):
    if type(x) == str:
        if x.split()[-1] == "km/kg":
            return float(x.split()[0]) * 1.40
        elif x.split()[-1] == "kmpl":
            return float(x.split()[0])
    else:
        return np.nan

df["Mileage [KMPL]"] = df["Mileage"].apply(mileage_convert)
```

```python
def power_convert(x):
    if type(x) == str:
        if x.split()[-1] == "bhp":
            return x.split()[0]
    else:
        return np.nan

df["Power_Capacity"] = df["Power"].apply(power_convert)
```

```python
def outlier_handling(x, p1= 0.25,p2 = 0.75):
    q1 = x.quantile(p1)
    q3 = x.quantile(p2)
    iqr = q3 - q1
    min_limit = q1 - 1.5 * iqr
    max_limit = q3 + 1.5 * iqr 
    return x[(x > min_limit) & (x < max_limit)] 
```

```python
for col in ["Engine_capacity","Power_Capacity","Mileage [KMPL]","Kilometers_Driven","Age"]:
    df2[col] = outlier_handling(df2[col])
    
```
